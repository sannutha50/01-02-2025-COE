```python
# Loop through all columns in the DataFrame
for column in df.columns:
    if df[column].dtype == 'object':  # Categorical column
        # Fill missing values with the mode of the column
        mode_value = df[column].mode()[0]  # Mode can return multiple values, so take the first one
        df[column].fillna(mode_value, inplace=True)
    elif df[column].dtype in ['int64', 'float64']:  # Numerical column
        # Fill missing values with the median of the column
        median_value = df[column].median()
        df[column].fillna(median_value, inplace=True)

# Check the DataFrame after filling missing values
print(df.isna().sum())  # This will show the number of missing values in each column
```

```python
# Removing HTML tags, special characters, and excessive whitespace.
import re
def clean_text(text):
    text = re.sub(r'<.*?>', '', text) 
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text) 
    text = re.sub(r'\s+', ' ', text).strip() 
    return text
df['Cleaned_JobDescription'] = df['JobDescription'].apply(clean_text)
df['Cleaned_JobRequirment'] = df['JobRequirment'].apply(clean_text)
df[['Cleaned_JobDescription', 'Cleaned_JobRequirment']].head()
```

```python
#Tokenizing and lowercasing for NLP tasks.
import re
def tokenize(text):
    tokens = text.lower().split()
    return ' '.join(tokens)
df['Cleaned_JobDescription'] = df['JobDescription'].apply(tokenize)
df['Cleaned_JobRequirment'] = df['JobRequirment'].apply(tokenize)
df[['Cleaned_JobDescription', 'Cleaned_JobRequirment']].head()
```

```python
#Normalize Title 
title_normalization = {
    "Software Eng.": "Software Engineer",
    "Senior Dev.": "Senior Developer",
    "Data Sci.": "Data Scientist",
    # Add more normalization rules as needed
}

# Function to normalize title
def normalize_title(title):
    return title_normalization.get(title, title)  # Return the normalized title or the original if not found

# Apply normalization to the 'Title' column
df['Title'] = df['Title'].apply(normalize_title)

# Display the normalized DataFrame
df['Title']
```

```python
#3.DATE OPERATIONS:
df['date'] = pd.to_datetime(df['date'], errors='coerce', dayfirst=True)  # Handle 'Jan 5, 2004' format
df['OpeningDate'] = pd.to_datetime(df['OpeningDate'], errors='coerce', dayfirst=True)  # Handle '17 July 2009' format
df['Deadline'] = pd.to_datetime(df['Deadline'], errors='coerce', dayfirst=True)  # Handle '26 January 2004' format

# The 'StartDate' column is not in a straightforward date format, so it might need custom handling.
# For now, we'll set it to NaT (Not a Time) as a placeholder. You may choose to handle such data differently.
df['StartDate'] = pd.to_datetime(df['StartDate'], errors='coerce')
df.dtypes
```

```python
# For StartDate column (custom handling needed based on your dataset)
# Here we just convert 'ASAP' and '3 months' to NaT (or you can fill it with specific dates as needed)
df['StartDate'] = pd.to_datetime(df['StartDate'], errors='coerce')
df['DaysLeftToApply'] = (df['Deadline'] - df['OpeningDate']).dt.days.abs()
df['DaysLeftToApply'].fillna(0,inplace=True)
df['JobStartDelay'] = (df['StartDate'] - df['OpeningDate']).dt.days.abs()
df['JobStartDelay'].fillna(0,inplace=True)
df['Year'] = df['OpeningDate'].dt.year
df['Month'] = df['OpeningDate'].dt.month

# Display the result
df[['DaysLeftToApply','JobStartDelay','Year','Month']]
```

```python
# Split Location into City and Country
df[['City', 'Country']] = df['Location'].str.split(',', expand=True, n=1)
df['City'] = df['City'].str.strip()
df['Country'] = df['Country'].str.strip()
df['Country'].fillna('Unknown', inplace=True)
df[[ 'City', 'Country']]
```
